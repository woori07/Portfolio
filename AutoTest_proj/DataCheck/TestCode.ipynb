```python
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.chrome.options import Options
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.common.by import By
from selenium.common.exceptions import NoSuchElementException
import time

# tr class 태그의 개수를 카운트하는 테스트 함수

class OpenBrowser:
    def startBrows(self):
        options = webdriver.ChromeOptions()
        options.add_experimental_option("detach", True)
        driver = webdriver.Chrome()
        driver.get('https://gs.thinkwise.co.kr/')
        driver.find_element(By.NAME, 'user_id').send_keys('secret')
        driver.find_element(By.NAME, 'user_pass').send_keys('secret')
        driver.find_element(By.ID, 'btn-confirm').click()
        alert = driver.switch_to.alert
        alert.accept()
        keyword = "회원관리"

        driver.find_element(By.XPATH, "//*[contains(text(),'" + keyword + "')]").click()
        time.sleep(3)
        options = Options()
        options.add_experimental_option("detach", True)
    
    
    
    # <tr class="tta_ctr"> 태그 개수를 담을 변수 선언 = countTr
    def tagReturn(self, countTr):

        # 첫 번째 페이지에서 <tr class="tta_ctr"> 태그 개수 찾기
        tr_tags = driver.find_elements(By.CSS_SELECTOR, 'tr.tta_ctr')
        countTr += len(tr_tags)

        # 다음 페이지로 계속 이동하며 태그 개수 카운팅
        while True:
            if driver.find_element(By.XPATH, '//*[@id="content-wrap"]/div/div[2]/a/img[1]').is_displayed():
                driver.find_element(By.XPATH, '//*[@id="content-wrap"]/div/div[2]/a/img[1]').click()
                tr_tags = driver.find_elements(By.CSS_SELECTOR, 'tr.tta_ctr')
                countTr += len(tr_tags)
                break

        # 찾은 태그의 수 출력
        print("찾은 <tr class='tta_ctr'> 태그의 수:", countTr)

     
if __name__=="__main__":
    openbrowser = OpenBrowser()
    openbrowser.startBrows()
    openbrowser.tagReturn(0)

    


```
